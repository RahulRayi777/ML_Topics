### SCALING AND CENTERING

### splitting into the x and y variable

```python
x=df.drop('DEFAULT',axis=1)
y=df['DEFAULT']
```

```python
df.dtypes
```

```python
encoded_x=pd.get_dummies(x,columns=['MARRIAGE','SEX','EDUCATION',
                                   'PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6'])
encoded_x
```

### THE RADIAL BASIS FUNCTION WE ARE USING OUR SUPPORT VECTOR M/C ASSUMES THE DATA ARE CENTERED AND SCALED ...SO, WE NEED TO DO THIS ON TRAININ  AND TESTING DATA

###  FIRST SPLIT THOSE

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(encoded_x,y,random_state=23)
```

```python
from sklearn.svm import SVC
model=SVC(random_state=23)
model.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import confusion_matrix
y_pred=model.predict(xtest)
cm=confusion_matrix(ytest,y_pred)
cm
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True)
```
